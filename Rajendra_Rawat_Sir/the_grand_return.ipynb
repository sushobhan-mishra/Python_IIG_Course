# Adding in list

## 1st location

```python
l = [1,2,3,4,5,6]
l.insert(0,0)
print(l)
```

## Specific location

```python
l = [1,2,3,4,5]
print(f'This my list: {l}')
i = int(input('Enter the Index you want to add:'))
n = int(input('Enter the number you want to add:'))

l.insert(i,n)
print(f'Added: {n}\nTo Index: {i}\nThis is the new list {l}')
```

```python
fruits = {'A':10,'B':15,'C':20,'D':13,'E':34,'F':78,'H':32}

A = list(fruits.keys())
print(len(A))

B = list(fruits.values())
print(max(B))
print(min(B))
```
