```python
'''11. Write a Python program using functions to find the value of nPr and nCr.'''

import math


def nPr(n, r):
    return math.factorial(n) // math.factorial(n - r)


def nCr(n, r):
    return math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

n, r = 5, 3
print(f"{n}P{r} =", nPr(n, r))
print(f"{n}C{r} =", nCr(n, r))
```

```python
'''12. Write a Python function named area that finds the area of a pentagon.'''

import math


def area_pentagon(side):
    # side and area rlation
    return (1 / 4) * math.sqrt(5 * (5 + 2 * math.sqrt(5))) * side ** 2

print("Area of pentagon:", area_pentagon(6))
```

```python
'''13. Write a program using functions to display Pascal's triangle.'''

from math import factorial

def pascal_triangle(n):
    for i in range(n):
        for j in range(n-i+1):
            '''Spacing'''
            print(end=" ")                                                               ######## check once more
            
        for j in range(i+1):
            '''Formula'''
            print(factorial(i)//(factorial(j)*factorial(i-j)), end=" ")

        # for new line
        print()

pascal_triangle(5)        
```

```python
'''14. Write a program using functions to print harmonic progression series and its sum till N terms.'''

def harmonic_progression(a,n,d):
    l = []
    for i in range(1,n+1):
        hp = 1 / ( a + ( (i - 1) * d ) )
        l.append(hp)
        s = sum(l)
    return l , s

# print(harmonic_progression(2,5,2))

HPn , Sn = harmonic_progression(2,5,2)
print("Harmonic Progression:", HPn)
print("Sum:", round(Sn, 5))
```

```python
'''17. Write a Python function that takes a string as input and returns a dictionary
containing the number of uppercase and lowercase characters in the string.'''

def count_case_chars(text):
    counts = {"uppercase": 0, "lowercase": 0}
    for ch in text:
        if ch.isupper():
            counts["uppercase"] += 1
        elif ch.islower():
            counts["lowercase"] += 1
    return counts

sample_text = "Hello Python 2026"
count_case_chars(sample_text)
```

```python
''' 18. Write a Python function that takes a list and an element as input.
 The function should add the element to the list only if it's not already present.'''

def add_if_not_present(lst, element):
    if element not in lst:
        lst.append(element)
        return True
    return False

items = [1, 2, 3]
print("Before:", items)

added = add_if_not_present(items, 4)
print( "After:", items)

added = add_if_not_present(items, 2)
print( "After:", items)
```
