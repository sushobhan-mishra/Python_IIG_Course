Swapping in 3 variable

```python
a=10
b=20
print('Before Swapping')
print(a)
print(b)
c=b
b=a
a=c
print('After Swapping')
print(a)
print(b)
```

Swapping in 2 variable using arithmetic operation

```python
a= 10
b=20
print('Before Swapping')
print(a)
print(b)
a=a+b
b=a-b
a=a-b
print('After Swapping')
print(a)
print(b)
```

Swapping in 2 variable using bitwise XOR

```python
a=10
b=20
print('Before Swapping')
print(a)
print(b)
a=a^b
b=a^b
a=a^b
print('After Swapping')
print(a)
print(b)
```
