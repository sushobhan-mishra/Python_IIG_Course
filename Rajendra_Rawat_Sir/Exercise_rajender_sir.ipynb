```python
n = int(input('Enter a number:'))
sum = 0
cnt = 0
for i in range(1,n + 1):
    if i % 2 == 0:
        sum = sum + i
        cnt = cnt + 1
Avg = sum / cnt
print(f'{sum}')
print(f'{cnt}')
print(f'Average of even numbers in {n} is {Avg:.2f}')
```

```python
n = int(input('Enter a number:'))
prod = 1
i = 1
while i < n+1:
    prod = prod * i
    i = i + 1     
print(f'Factorial of {n} is {prod}.')    
```

```python
n = int(input('Enter a number:'))
prod = 1
i = 1
cnt = 0
print(f'0! = 1.')
while i < n+1:
    prod = prod * i
    i = i + 1     
    cnt = cnt + 1
    print(f'{cnt}! = {prod}.')
```

```python
stud_name = input('Enter the student name:')
s_1,s_2,s_3 = input('Enter three subjects with space:').split(' ')
m_1 = int(input(f'Enter marks of {s_1}:'))
m_2 = int(input(f'Enter marks of {s_2}:'))
m_3 = int(input(f'Enter marks of {s_3}:'))
print('MARKSHEET')
print(f'{s_1} = {m_1}\n{s_2} = {m_2}\n{s_3} = {m_3}')
print()
tot_mrk = 300
avg = (m_1 + m_2 + m_3) / tot_mrk                   #Average
per = avg * 100                                     #Percentage
if per <= 40:
    print(f'Percentage of {stud_name} is {per:.2f}%')
    print('Fail\nPhir milte hain😉')
elif 40 < per < 50:
    print(f'Percentage of {stud_name} is {per:.2f}%')
    print('Grade: D\nIss baar bachh gayee😑') 
elif 50 <= per <= 60:
    print(f'Percentage of {stud_name} is {per:.2f}%')
    print('Grade: C\nChiiiii🤢')
elif 60 <= per <= 70:
    print(f'Percentage of {stud_name} is {per:.2f}%')
    print('Grade: B\nEwwww🤮')
elif 70 <= per <= 80:
    print(f'Percentage of {stud_name} is {per:.2f}%')
elif 80 <= per <= 90:
    print(f'Percentage of {stud_name} is {per:.2f}%')
    print('Grade: A\nThik he chalega🤧')
else:
    print(f'Percentage of {stud_name} is {per:.2f}%')
    print('Grade: A+\nMene padhaya hai tumhe🥰')                   
```

```python
print('                     SBI ATM                       ')
print('1. Balance Check\n2. Deposit\n3. Withdrawl\n4. Exit')
s = int(input('Enter the required number:'))
avl_amt = 0
while s < 4:
    s = int(input('Enter the required number:'))
    if s == 1:
        print('                     SBI ATM                       ')
        print('1. Balance Check\n2. Deposit\n3. Withdrawl\n4. Exit')
        print(f'\nYour Account Balance is {avl_amt}')
    elif s == 2:
        print('                     SBI ATM                       ')
        print('1. Balance Check\n2. Deposit\n3. Withdrawl\n4. Exit')
        n = int(input('Enter the deposition amount:'))
        avl_amt += n
        print(f'\nAvailable Amount: {avl_amt}')
    elif s == 3:
        print('                     SBI ATM                       ')
        print('1. Balance Check\n2. Deposit\n3. Withdrawl\n4. Exit')
        n = int(input('Enter the withdrawl amount:'))
        avl_amt -= n
        print(f'\nAvailable Amount: {avl_amt}') 
print('\nTHANK YOU FOR BANKING WITH US.')

```

```python
n = int(input('Enter a number:'))
reversed_num = 0
while n > 0:
       digit = n % 10           # get last digit
       reversed_num = reversed_num * 10 + digit            #shift and add digits
       n = n // 10            # remove last digit
print(reversed_num)
```
