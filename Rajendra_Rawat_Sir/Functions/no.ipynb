```python
def monthly_payment(loan_amount, years, annual_interest_rate):
    monthly_rate = annual_interest_rate / 1200
    months = years * 12
    if monthly_rate == 0:
        return loan_amount / months
    return loan_amount * monthly_rate / (1 - (1 + monthly_rate) ** -months)

print("Monthly payment for loan 100000, 5 years, 7%:", round(monthly_payment(100000, 5, 7), 2))
```
