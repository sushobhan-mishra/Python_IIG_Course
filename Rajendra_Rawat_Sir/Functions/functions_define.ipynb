```python
def decimal_to_binary(n):
    if n == '0':
        print('0')
    binary =''    
    while n > 0:
        binary = str(n % 2) + binary       # get the remiander
        n = n//2                           # remove the remainder
    return binary
decimal_to_binary(10)    
```

```python
def fact(n):
    if n <= 1:
        return(1)
    else:
        return n * fact(n-1)
fact(5)    
```

```python
def countdown(n):
    if n <= 0:
        print('Boom!')
    else:
        print(n)
        countdown(n - 1)
countdown(11)                   
```

```python
def countdown(n):
    if n <= 0:
        print('Boom!')
    elif n % 2 != 0:
        print(n)
        countdown(n-1)
    elif n % 2 == 0:
        countdown(n-1)
countdown(11) 
```

```python
import math
def permutation(n,r):
    nPr = math.factorial(n)/math.factorial(n - r)
    return nPr
print(permutation(5,4))
def combination(n,r):
    nCr = math.factorial(n)/(math.factorial(r)*math.factorial(n - r))
    return nCr
combination(5,4)
```
