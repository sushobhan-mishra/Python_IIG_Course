```python
x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a,b : a*b
print(x(5,0))
```

```python

x = lambda a,b,c,d : a + b + c + d
print(x(5,6,7,8))
```

```python

check = lambda a : 'Even' if a % 2 == 0 else 'Odd'
print(check(4))
print(check(7))
```

```python
check = lambda a : f'{a} divisible by 3' if a % 3 == 0 else f'{a} not divisible by 3'
print(check(3))
print(check(7))
```

```python
check = lambda a,b : 'a is greater than b' if a > b  else 'b is greater than a'
print(check(4,10))
```
