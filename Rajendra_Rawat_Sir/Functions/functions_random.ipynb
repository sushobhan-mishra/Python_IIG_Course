```python
print('1. Absolute value\n2. Mod division\n3. Power \n4. length of a string\n5. Exit')
n = int(input('Enter a number for required function:'))
while n < 5:
    if n == 1:
        A = int(input('Enter the number for absolute value:'))
        print(abs(A))
        n = int(input('Enter a number for required function:'))
    elif n == 2:
        B = int(input('Enter the dividend for Mod division:'))
        C = int(input('Enter the divisor for Mod division:'))
        print(divmod(B,C))
        n = int(input('Enter a number for required function:'))
    elif n == 3:
        D = int(input('Enter the number for Power:'))
        E = int(input('Enter the power:'))  
        print(pow(D,E))
        n = int(input('Enter a number for required function:'))
    elif n ==4:
        D = input('Enter the string for length:')
        print(len(D))
        n = int(input('Enter a number for required function:'))
else:
    print('Exit')        

```
