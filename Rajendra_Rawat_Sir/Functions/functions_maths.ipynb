1.Trigometric Function    (output in radian)

```python
print(math.cos(0))
print(math.sin(0))
print(math.tan(0))
print(math.asin(0))            #Sin inverse 
print(math.acos(0))
print(math.atan(0))
print(math.atan2(34,-7))       # between -π and π.to determine the correct quadrant of the angle. involving vector calculations
```

2.Power, exponential and logarithmic functions

```python
print(math.cbrt(16)) # cuberoot
print(math.exp(0))     #e to power x
print(math.exp2(2))   # 2 to power x
print(math.expm1(0))  # e to power x minus 1
print(math.log(1,10)) #log of x to base(e by default)
print(math.log1p(0)) #log 1+x base e
print(math.log2(1)) #base 2 log x
print(math.log10(1))
print(math.pow(3,2))
print(math.sqrt(4))
```
