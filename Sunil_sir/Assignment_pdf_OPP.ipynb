Question 1

```python
# Create a clas
class Person:
    # write __init__ method
    def __init__(self,name,age):
        self.name = name
        self.age = age

    # write greet method    
    def greet(self):
        return f'Hello, my name is {self.name}'
#create an object
p1 = Person('John', 36)

#Call the greet method
p1.greet()
```

```python
class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def bark(self):
        return f'{self.name} says Woof!'

d1 = Dog('Buddy',3)

d1.bark()
```

Question 5

```python
class Rectangle:
    def __init__(self,width,height):
        self.width = width
        self.height = height

    def area(self):
        return f'Area of Rectangle having width {self.width} cm and height {self.height} cm is {self.width * self .height}'

r1 = Rectangle(5,3)

r1.area()
```
