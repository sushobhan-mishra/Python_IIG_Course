List is Ordered, Changeable and can be duplicated.

```python
thislist = ['apple','banana','cherry']
print(thislist)
print(len(thislist)) 
print(type(thislist)) 
```

```python
thislist = ["apple", "banana", "cherry", "apple", "cherry"] 
print(thislist)
```

List constructor

```python
thislist = list(("apple", "banana", "cherry")) # note the double round-brackets 
print(thislist)
```

```python
thislist = ["apple", "banana", "cherry"] 
print(thislist[1]) 
print(thislist[-1]) 

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"] 
print(thislist[2:5])
print(thislist[:4]) 
print(thislist[2:]) 
print(thislist[-4:-1]) 
```

Checking the existence of an item

```python
thislist = ["apple", "banana", "cherry"] 
if "apple" in thislist: 
  print("Yes, 'apple' is in the fruits list") 
```

changing items in list

```python
thislist = ["apple", "banana", "cherry"] 
thislist[1] = "blackcurrant" 
print(thislist)

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"] 
thislist[1:3] = ["blackcurrant", "watermelon"] 
print(thislist)

thislist = ["apple", "banana", "cherry"] 
thislist[1:2] = ["blackcurrant", "watermelon"] 
print(thislist) 

thislist = ["apple", "banana", "cherry"] 
thislist[1:3] = ["watermelon"] 
print(thislist) 
```

Add List Items

```python
thislist = ["apple", "banana", "cherry"] 
thislist.append("orange") 
print(thislist)
```

```python
thislist = ["apple", "banana", "cherry"] 
thislist.insert(1, "orange") 
print(thislist)
```

```python
thislist = ["apple", "banana", "cherry"] 
tropical = ["mango", "pineapple", "papaya"] 
thislist.extend(tropical) 
print(thislist)
```

```python
thislist = ["apple", "banana", "cherry"] 
thistuple = ("kiwi", "orange") 
thislist.extend(thistuple) 
print(thislist) 
```

Remove List Items

```python
thislist = ["apple", "banana", "cherry"] 
thislist.remove("banana") 
print(thislist) 

thislist = ["apple", "banana", "cherry", "banana", "kiwi"] 
thislist.remove("banana") 
print(thislist) 
```

```python
thislist = ["apple", "banana", "cherry"] 
thislist.pop(1) 
print(thislist) 

thislist = ["apple", "banana", "cherry"] 
thislist.pop() 
print(thislist) 
```

```python
thislist = ["apple", "banana", "cherry"] 
del thislist[0] 
print(thislist) 

thislist = ["apple", "banana", "cherry"] 
del thislist 
```

```python
thislist = ["apple", "banana", "cherry"] 
thislist.clear() 
print(thislist) 
```

Loop Lists

```python
thislist = ["apple", "banana", "cherry"] 
for x in thislist:
    print(x)
```

```python
thislist = ["apple", "banana", "cherry"] 
for i in range(len(thislist)): 
   if i == 1:
     print(thislist[i])
```

```python
thislist = ["apple", "banana", "cherry"] 
i = 0
while i < len(thislist):
    if i == 2:
      print(thislist[i])
    i = i + 1
```

Without list comprehension

```python
fruits = ['apple','banana','cherry','kiwi','mango']
newlist = []

for x in fruits:
    if 'a' in x:
        newlist.append(x)
print(newlist)        
```

With list comprehension

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"] 
newlist = [x for x in fruits if "a" not in x] 
print(newlist)
```

```python
fruits = ["apple", 'banana', "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if x != 'apple']
print(newlist)
```

Sort List

```python
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"] 
thislist.sort() 
print(thislist) 
```

```python
thislist = [100, 50, 65, 82, 23] 
thislist.sort() 
print(thislist) 
```

```python
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort(reverse = True)
print(thislist)
```

```python
thislist = [100, 50, 65, 82, 23] 
thislist.sort(reverse = True) 
print(thislist) 
```

Customize sort function

```python
def myfunc(n):
    return abs(n-50)

thislist = [100,50,65,82,23]
thislist.sort(key = myfunc)
print(thislist)
```

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"] 
thislist.sort(key = str.lower) 
print(thislist)
```

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"] 
thislist.reverse() 
print(thislist) 
```
