Question 1: Single Integration of a Polynomial Function

```python
from scipy.integrate import quad

def f(x):
    return (x**2) + 5*x + 6

result,error = quad(f, 0 , 2)

print(f"({result},{error})")
```

Question 2: Single Integration of a Trigonometric Function 

```python
from scipy.integrate import quad
import numpy as np

result,error = quad(lambda x:np.sin(x)**2, 0 , np.pi/2)

print(f"({result},{error})")
```

Question 3: Single Integration using a Lambda Function

```python
from scipy.integrate import quad

result,error = quad(lambda x:x**3 + 2*x , 1 , 3)

print(f"({result},{error})")
```

Question 4: Single Integration of an Logarithmic Function

```python
from scipy.integrate import quad
import numpy as np

result,error = quad(np.log, 0 , 4)

print(f"({result},{error})")
```

Question 5: Double Integration of a Polynomial Function 

```python
from scipy.integrate import dblquad 

f = lambda y, x: x**2 + y**2 

result, error = dblquad( f, 0, 2, lambda x: 0, lambda x: 1 ) 

print(result)
```

Question 6: Double Integration for Area Calculation 

```python
from scipy.integrate import dblquad 

f = lambda y, x: x + y

result, error = dblquad( f, 0, 3, lambda x: 0, lambda x: 2 ) 

print(result)
```
