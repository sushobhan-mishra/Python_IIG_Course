```python
import math as mt 
```

```python
class Shape:
    def __init__(self, shape_name):
        self.shape_name = shape_name

    def display(self):
        pass                                  
```

```python
class Shape2D(Shape):
    def __init__(self):
        pass

    def calculate_area(self):
        pass

    def calculate_perimeter(self):
        pass
```

2D

```python

class Circle(Shape2D):
    def __init__(self, radius):
        self.__radius = radius

    def calculate_area(self):
        return f'Area of Circle: {mt.pi*(self.__radius**2)}'

    def calculate_perimeter(self):
        return f'Circumference of Circle: {2*mt.pi*self.__radius}'   
##########################################################
class Rectangle(Shape2D):
    def __init__(self, length, width):
        self.__length = length
        self.__width = width
      
    

    def calculate_area(self):
        return f'Area of Rectangle: {self.__length * self.__width}'

    def calculate_perimeter(self):
        return f'Perimeter of Rectangle: {2 * (self.__length + self.__width)}'   

class Square(Rectangle):
    def __init__(self, side):
        #self.__side = side
        super().__init__(side,side)
        self.shape_name ='Square'

    ''' def calculate_area(self):
        return f'Area of Square: {self.__side**2}'

    def calculate_perimeter(self):
        return f'Perimeter of Square: {4 * self.__side}'  ''' 
#############################################################
class Triangle(Shape2D):
    def __init__(self, base, height, side1, side2, side3):
        self.__base = base
        self.__height = height
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def calculate_area(self):
        return f'Area of Circle: {0.5 * self.__base * self.__height}'

    def calculate_perimeter(self):
        return f'Perimeter of Circle: {self.__side1 + self.__side2 + self.__side3}'   
```

```python
shapes = [ 
    Circle(5), 
    Rectangle(10, 4), 
    Square(6), 
    Triangle(4, 5, 3, 4, 5) 
] 
 
for shape in shapes: 
    print(shape.calculate_area()) 
    print(shape.calculate_perimeter())  
```

3D

```python
# 3D
class Shape3D(Shape):
    def __init__(self):
        pass

    def calculate_volume(self):
        pass

class Cone(Shape3D):
    def __init__(self, radius, height):
        self.__radius = radius
        self.__height = height

    def calculate_volume(self):
        return f'Volume of Cone: {(1 * mt.pi * (self.__radius**2) * self.__height)/3}'

class Cylinder(Shape3D):
    def __init__(self, radius, height):
        self.__radius = radius
        self.__height = height 

    def calculate_volume(self):
        return f'Volume of Cylinder: {mt.pi * (self.__radius**2) * self.__height}'    

class Sphere(Shape3D):
    def __init__(self, radius):
        self.__radius = radius

    def calculate_volume(self):
        return f'Volume of Sphere: {(4 * mt.pi * (self.__radius**3))/3}'    
####################################################
class Cuboid(Shape3D):
    def __init__(self, length, width, height):
        self.__length = length
        self.__width = width
        self.__height = height 

    def calculate_volume(self):
        return f'Volume of Cuboid: {self.__length * self.__width * self.__height}'    

class Cube(Cuboid):
    def __init__(self, side):
        self.__side = side

    def calculate_volume(self):
        return f'Volume of Cube: {self.__side**3}'   
```

```python
solids = [ 
Cone(3, 5), 
Cylinder(3, 5), 
Sphere(4), 
Cube(6) 
] 

for solid in solids: 
    print(solid.calculate_volume()) 
```
