## Exercise 1: `point_str`

The first step in defining most classes is to define their `__init__` and `__str__` methods so that we can construct and represent distinct objects of that class. Our `Point` class should accept two arguments, `x` and `y`, and be represented by a string `'Point(x, y)'` with appropriate values for `x` and `y`.


```python
# Represents a point in a 2-dimensional coordinate system.
class Point:
    #Initialize a Point object with x and y coordinates.
    def __init__(self, x, y):
        self.x = x
        self.y = y

    #Return a human-readable string representation of the Point.
    def __str__(self):
        return f"Point({self.x}, {self.y})"
```

```python
p=Point(1,2)
print(p)
```

## Exercise 2: add_subtract

The most basic vector operations we want our `Point` object to handle are addition and subtraction. For two points $(x_1, y_1) + (x_2, y_2) = (x_1 + x_2, y_1 + y_2)$ and similarly for subtraction. Implement a method within `Point` that allows two `Point` objects to be added together using the `+` operator, and likewise for subtraction.

(Remember that `__add__` and `__sub__` methods will allow us to use the `+` and `-` operators.)

```python
# Write your code here
class Point:
    #Initialize a Point object with x and y coordinates.
    def __init__(self, x, y):
        self.x = x
        self.y = y

    
    def __add__(self,other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __sub__(self,other):
        return Point(self.x - other.x, self.y - other.y)

    def __str__(self):
        return f'Points ({self.x},{self.y})'
    
    
 
```

```python
# Execute and Test
p1 = Point(1,2) 
p2 = Point(3,4) 
p3 = Point(5,4) 
print(p1)
print(p2)
print(p3)
print()
print(p1.x)
print(p1.y)
print(p1.__add__(p2))
print(p1 + p2 )        #python automatically calls p1.__add__(p2)
print(p1-p2)          #python automatically calls p1.__sub__(p2)
```

## Exercise 3: multiplication

Within linear algebra there's many different kinds of multiplication: scalar multiplication, inner product, cross product, and matrix product. We're going to implement scalar multiplication and the inner product.

We can define scalar multiplication given a point $P$ and a scalar $a$ as
$$aP=a(x,y)=(ax,ay)$$

and we can define the inner product for points $P,Q$ as
$$P\cdot Q=(x_1,y_1)\cdot (x_2, y_2) = x_1x_2 + y_1y_2$$


(Remember that `__mul__` method will allow us to use the `*` operator. Also don't forget that the ordering of operands matters when implementing these operators.)

```python
# Write your code here
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self,other):
        if isinstance(other,Point):
            return self.x * other.x + self.y * other.y
        if isinstance(other,(int,float)):
            return Point(other*self.x, other*self.y)

    def __str__(self):
        return f'Point({self.x}, {self.y})'        
```

```python
# Execute and Test
p1 = Point(1,2)
p2 = Point(3,4)
p3 = p1*p2
print(p3)
print(p1*3)
```

## Exercise 4: Distance

Another quantity we might want to compute is the distance between two points.  This is generally given for points $P_1=(x_1,y_1)$ and $P_2=(x_2,y_2)$ as
$$D = |P_2 - P_1| = \sqrt{(x_1-x_2)^2 + (y_1-y_2)^2}.$$

Implement a method called `distance` which finds the distance from a point to another point.

```python
# Write your code here
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def dist(self, other):
        return((self.x - other.x)**2 + (self.y - other.y)**2)**0.5    
```

```python
# Execute and Test
p1 = Point(1,2)
p2 = Point(3,4)
p1.dist(p2)
```
