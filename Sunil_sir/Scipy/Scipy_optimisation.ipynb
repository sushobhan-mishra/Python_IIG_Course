```python
from scipy.optimize import root 
from numpy import cos 
```

Roots of an Equation

```python
def eqn(x): 
    return x + cos(x) 

myroot = root(eqn, 0)  # root(fun,x0)

# fun --> a function representing an equation
# x0 --> an initial guess for the root
 
print(myroot.x)
```

```python
print(myroot)  # x is the actual value
```

```python
print(myroot.x)    # solution
print(myroot.success)
print(myroot.message)
print(myroot.fun)
```

Minimizing  function

In a Curve,High points are called maxima. 
Low points are called minima. 

The highest point in the whole curve is called global maxima, whereas the rest of them 
are called local maxima. 
The lowest point in whole curve is called global minima, whereas the rest of them are 
called local minima. 

```python

# minimize(fun,x0,method,callback,options)
# method -->  name of the method to use. Legal values: 
                                    # 'CG' 
                                    # 'BFGS' 
                                    # 'Newton-CG' 
                                    # 'L-BFGS-B' 
                                    # 'TNC' 
                                    # 'COBYLA' 
                                    # 'SLSQP'

#callback --> function called after each iteration of optimization
# options --> a dictionary defining extra params:
        #    { "disp": boolean - print detailed description 
            #  "gtol": number - the tolerance of the error } 
```

```python
from scipy.optimize import minimize 
def eqn(x): 
    return x**2 + x + 2 
mymin = minimize(eqn, 0, method='BFGS') 
print("Minimum occurs at x =", mymin.x[0]) 
print("Minimum value =", mymin.fun) 
```

```python
from scipy.optimize import minimize 

def cost(x): 
    return (x - 5)**2 + 10 

result = minimize(cost, x0=0) 
print("Minimum occurs at x =", result.x) 
print("Minimum value =", result.fun) 
```
