Definite integral

Single Integration using quad() 

```python
from scipy.integrate import quad 

def f(x): 
    return x**2 

result, error = quad(f, 0, 1) 

print("Result =", result)  
print("Error =", error)
```

Integrating Trigonometric Functions 

```python
from scipy.integrate import quad 
import numpy as np 

# also works in math module
result, error = quad(np.sin, 0, np.pi) 

print(result) 
print(error)
```

Integrating Exponential Functions

```python
from scipy.integrate import quad 
import numpy as np

result, error = quad( lambda x: np.exp(-x), 0, 5) 
print(result)
```

Infinite Integrals 

```python
from scipy.integrate import quad 
import numpy as np 

result, error = quad(lambda x: np.exp(-x), 0, np.inf ) 
print(result)
```

Double Integration

```python
from scipy.integrate import dblquad 

f = lambda y, x: x*y 

result, error = dblquad( f, 0, 1, lambda x: 0, lambda x: 1 ) 

print(result)
```
