OOP encapsulation(process of converting instance variable to private or protected variable by using _or__ respectively)

```python
class Alibag_Employee:
    Headquarter = 'Indian Institute of Geomagnetism'
    Place_headquarter = 'Panvel'
    Observatory = 'Magnetic Observatory Alibag'
    Place_observatory = 'Alibag'
    District = 'Raigad'                                               # class variable for fixed value
    State = 'Maharastra'
    HRA = 15
    TA = 10

    def __init__(self,name,post,pay_level,age,place):

        self.name = name
        self.post = post
        self.pay_level = pay_level                                      # instance variable is for varried value
        self.age = age
        self.place = place
        self._basic_pay = 44900 # private variable(_)           # for developer
        self.__DA = 15 # protected variable(__)
        self.__password = 7856

    def monthly_net_salary(self):
        print(f'Net monthly salary of {self.name} is {self._basic_pay + (self.__DA/100)*(self._basic_pay) + (self.HRA/100)*(self._basic_pay) + (self.TA/100)*(self._basic_pay)}.')

    def show_da(self):
        self.input_password = int(input("Enter 4 digit PIN: "))
        if self.input_password == self.__password:
            print(f"{self.__DA}")
employee1 = Alibag_Employee('Pranjal Saikia','Scientific Officer','7','34','Assam')
employee1.monthly_net_salary()
print(employee1.place) # instance variable can be accesed from object
employee1.show_da()



```

```python
class UCO_bank():
    

    def __init__(self,name,address,tot_bal,deposit,withdrawl):
        
        self.name = input('Enter your name')
        self.address = input(f'Address of {self.name}')
        self.tot_bal = tot_bal
        self.deposit = deposit
        self.withdrawl = withdrawl

    def profile(self):
        print('Name:',self.name.upper())
        print('Address:',self.address.upper())

    def process(self):
        print('1.Total Balance\n2.Deposit\n3.Withdrawl\n4.Exit')
        i = int(input('Enter the required number'))
        while i < 5:
            if i == 1:
                print('Available amount:',self.tot_bal)
                print('\n1.Total Balance\n2.Deposit\n3.Withdrawl\n4.Exit')
                i = int(input('Enter the required number'))
            elif i == 2:
                self.deposit = int(input('Enter the deposit amount'))
                self.tot_bal = int(self.deposit) + int(self.tot_bal)
                print(f'\nAmount Deposited: {self.deposit}')
                print('\n1.Total Balance\n2.Deposit\n3.Withdrawl\n4.Exit')
                i = int(input('Enter the required number'))
            elif i == 3:
                self.withdrawl = int(input('Enter the Withdrawl amount'))
                self.tot_bal = int(self.tot_bal) - int(self.withdrawl)
                print(f'\nAmount Withdrawn: {self.withdrawl}') 
                print('\n1.Total Balance\n2.Deposit\n3.Withdrawl\n4.Exit')
                i = int(input('Enter the required number'))
            elif i == 4:
                print('\nTHANK YOU FOR BANKING WITH US😁.')
                break    
            else:
                print('Wrong Input')              
user = UCO_bank('','',0,'','')
user.profile()
user.process()        
```
