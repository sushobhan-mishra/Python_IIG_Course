```python
# parent class
class Alibag_Employee:
    Headquarter = 'Indian Institute of Geomagnetism'
    Place_headquarter = 'Panvel'
    Observatory = 'Magnetic Observatory Alibag'
    Place_observatory = 'Alibag'
    District = 'Raigad'                                               # class variable for fixed value
    State = 'Maharastra'
    HRA = 15
    TA = 10

    def __init__(self,name,post,pay_level,age,place):

        self.name = name
        self.post = post
        self.pay_level = pay_level                                      # instance variable is for varried value
        self.age = age
        self.place = place
        self._basic_pay = 44900 # private variable(_)           # for developer
        self.__DA = 15 # protected variable(__)
        self.__password = 7856

    def monthly_net_salary(self):
        print(f'Net monthly salary of {self.name} is {self._basic_pay + (self.__DA/100)*(self._basic_pay) + (self.HRA/100)*(self._basic_pay) + (self.TA/100)*(self._basic_pay)}.')

    def show_da(self):
        self.input_password = int(input("Enter 4 digit PIN: "))
        if self.input_password == self.__password:
            print(f"DA: {self.__DA}%")

employee1 = Alibag_Employee('Pranjal Saikia','Scientific Officer','7','34','Assam')

# child class
class IIGtech(Alibag_Employee):

    def __init__(self,name,post,pay_level,age,place,pin,techgroup,project,techwork,patent):

        super().__init__(name,post,pay_level,age,place)

        self.pin = pin
        self.techgroup = techgroup
        self.project = project
        self.techwork = techwork
        self.patent = patent

    def technical_info(self):
        print(self.techgroup)    
        print(self.project) 
        print(self.techwork) 
        print(self.patent) 

#tech1 and employee is object and also called instance

tech1 = IIGtech('Sushobhan','Technical Officer','10','35','Bhubaneswar','Instrumention','Data analysis','XYZ','fabrication','2')

# instance method
tech1.technical_info()
tech1.monthly_net_salary()
tech1.show_da()

employee1.monthly_net_salary()
employee1.show_da()
print(employee1.place) # instance variable can be accesed from object
print(tech1.place)

# parent object cannot call child method
# but child onject can call parent method

# #class method
# @classmethod

# #static method
# @staticmethod

```
