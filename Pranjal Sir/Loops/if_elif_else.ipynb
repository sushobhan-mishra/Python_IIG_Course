1.WAP to check if a number is positive.

```python
n = int(input('Enter a number:'))
if n > 0:
    print(f'{n} is positive')
elif n == 0:
    print(f'{n}')
else:
    print(f'{n} is negative')        
```

2.Check if a number is a even

```python
n = int(input('Enter a number:'))
print(n)
print('You entered number:', {n},type(n)) 
if n % 2 == 0:
    print(f'{n} is even')
else:
    print(f'{n} is odd') 
    
```

3.Find the largest of 3 number

```python
# a = int(input('a:'))
# b = int(input('b:'))
# c = int(input('c:'))
a,b,c = map(int, input('Enter three numbers with space: ').split(' '))
print(a,b,c)
print('You entered the first number a:',type(a))
print('You entered the first number b:',type(b))
print('You entered the first number c:',type(c))
if a > b and a > c:
    print(f'{a} is largest')
elif b > a and b > c:
    print(f'{b} is largest')
else:
    print(f'{c} is largest')  
```
