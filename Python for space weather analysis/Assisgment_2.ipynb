```python
year = []
doy = []
hour = []
minute =[]

symh = []

file = open("C:\\Users\\mo alibag\\Desktop\\IIG Sushobhan Mishra\\Python for space weather analysis\\7_8_sep_2017.txt","r")

for line in file:

    values = line.split()
    try:
        year.append(int(values[0]))
        doy.append(int(values[1]))
        hour.append(int(values[2]))
        minute.append(int(values[3]))

         # OMNI parameters
        symh.append(float(values[4])) 
    except:

        continue  
file.close()       
```

```python
import numpy as np

symh = np.array(symh)

symh[symh > 999] = np.nan
```

```python
import matplotlib.pyplot as plt      # plt is variable

# create figure
plt.figure(figsize=(12,8))

# Bz nT
plt.subplot(5,1,1)
plt.plot(symh)

plt.ylabel('symh (nT)')
plt.xlabel('IMF Bz')
```

```python
minimum_symh = min(symh)

if minimum_symh < -250:
    print('Super Strom')

elif minimum_symh < -100:
    print('Intense Strom')

elif minimum_symh < -50:
    print('Moderate Strom')

elif minimum_symh < -10:
    print('Weak Disturbance')
else:
    print('No strom') 
```

```python
def classify_strom(data['symH']):
    minimum_symh = data['symH'].min()

    if minimum_symh < -250:
      return 'Super Strom'

    elif minimum_symh < -100:
      return 'Intense Strom'

    elif minimum_symh < -50:
      return 'Moderate Strom'

    elif minimum_symh < -10:
      return 'Weak Disturbance'
    else:
      return 'No strom'
    
classify_strom(-20)   
```
