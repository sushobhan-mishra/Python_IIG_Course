```python
import pandas as pd
data = pd.read_csv(r"C:\Users\mo alibag\Desktop\IIG Sushobhan Mishra\Python for space weather analysis\16_dec_2003_2004.txt",
                    sep = r'\s+',
                    skiprows = 6,
                    skipfooter = 3,
                    engine = 'python'
                   )
```

```python
print(data.head(10))
```

```python
data.columns = [
    'Year',
    'DOY',
    'Hour',
    'Minute',
    'symH'
]
```

```python
print(data.head(10))
```

```python
days = []

for i in range(len(data['symH'])):

    if data['symH'][i] < -30:
        days.append(int(data['DOY'][i]))
        
# unique distributed days
unique_days = set(days)

print(unique_days)
```
