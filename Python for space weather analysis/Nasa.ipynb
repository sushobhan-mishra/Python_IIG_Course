Assinging data in list

```python
year = []
doy = []
hour = []
minute =[]

bz = []
speed = []
density = []
pressure = []
symh = []

file = open("C:\\Users\\mo alibag\\Desktop\\IIG Sushobhan Mishra\\Solar_wind.txt","r")

for line in file:

    values = line.split()
    try:
        year.append(int(values[0]))
        doy.append(int(values[1]))
        hour.append(int(values[2]))
        minute.append(int(values[3]))

         # OMNI parameters
        bz.append(float(values[4]))
        speed.append(float(values[5]))
        density.append(float(values[6]))
        symh.append(float(values[7])) 
    except:

        continue  
file.close()       
```

To plot the graph

```python
import matplotlib.pyplot as plt      # plt is variable

# create figure
plt.figure(figsize=(12,8))

# Bz nT
plt.subplot(5,1,1)
plt.plot(bz)

plt.ylabel('Bz (nT)')
plt.xlabel('IMF Bz')
```

To clean the data

```python
import numpy as np

bz = np.array(bz)
speed = np.array(speed)
density = np.array(density)
pressure = np.array(pressure)
symh = np.array(symh)

bz[bz > 999] = np.nan

speed[speed > 999] = np.nan

density[density > 999] = np.nan

pressure[pressure > 999] = np.nan
```

Analysis of the Earth's magnetic field variation in term of Solar storm

```python
minimum_symh = min(symh)

if minimum_symh < -250:
    print('Super Strom')

elif minimum_symh < -100:
    print('Intense Strom')

elif minimum_symh < -50:
    print('Moderate Strom')

elif minimum_symh < -10:
    print('Weak Disturbance')
else:
    print('No strom')                
```
