Assigning in list

```python
year = []
doy = []
hour = []
minute =[]

symh = []

file = open("C:\\Users\\mo alibag\\Desktop\\IIG Sushobhan Mishra\\Python for space weather analysis\\17_18_mar_2015.txt","r")

for line in file:

    values = line.split()
    try:
        year.append(int(values[0]))
        doy.append(int(values[1]))
        hour.append(int(values[2]))
        minute.append(int(values[3]))

         # OMNI parameters
        symh.append(float(values[4])) 
    except:

        continue  
file.close()       
```

Cleaning the data

```python
import numpy as np

symh = np.array(symh)

symh[symh > 999] = np.nan
```

Plotting the data

```python
import matplotlib.pyplot as plt      # plt is variable

# create figure
plt.figure(figsize=(12,8))

# Bz nT
plt.subplot(5,1,1)
plt.plot(symh)

plt.ylabel('symh (nT)')
plt.xlabel('IMF Bz')
```

Analysis of data

```python
minimum_symh = min(symh)

if minimum_symh < -250:
    print('Super Strom')

elif minimum_symh < -100:
    print('Intense Strom')

elif minimum_symh < -50:
    print('Moderate Strom')

elif minimum_symh < -10:
    print('Weak Disturbance')
else:
    print('No strom') 
```
