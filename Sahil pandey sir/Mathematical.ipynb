```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5,1000)
a = 0
b = 0
y = np.sqrt(25-((x**2)- a)) + b
plt.plot(x,y,c='red',linewidth=2)
plt.plot(x,-y,c='red',linewidth=2)
plt.axis('equal')

# diameter line
  # [5,-5]= x    [0,0]=y
plt.plot([-5,5],[0,0], c='red',linewidth=2)
plt.plot([0,0],[5,-5], c='red',linewidth=2)

# diagonal line
plt.plot([-5/np.sqrt(2),5/np.sqrt(2)],[-5/np.sqrt(2),5/np.sqrt(2)], c='red',linewidth=2)
plt.plot([-5/np.sqrt(2),5/np.sqrt(2)],[5/np.sqrt(2),-5/np.sqrt(2)], c='red',linewidth=2)

# center point
plt.plot(0,0, 'o',markersize = 5, c = 'blue') 
plt.grid()
plt.show()
```
