```python
# Example 2

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 360.5, 2)
#x =  np.linspace(0, 360, 71) # 0 to 365 in 100 evenly spaced
y1 = np.sin(np.deg2rad(x)); 
y2 = np.sin(2*np.deg2rad(x)); 
y3 = np.cos(np.deg2rad(x)); 
y4 = np.cos(2*np.deg2rad(x)); 

#print(x)  # python(y)
plt.plot(x, y1,'-b', linewidth = '0.5',label='sin x')
plt.plot(x, y2,'--b',linewidth = '0.5',label='sin 2x')
plt.plot(x, y3,'.-r',linewidth = '0.5', label='cos x')
plt.plot(x, y4,'-.r',linewidth = '0.5', label='cos 2x')
plt.title("Sine and cosine waves")
plt.xlabel("x\N{DEGREE SIGN} ")
plt.ylabel("Amplitude")
plt.xlim(0, 360)
plt.ylim(-1.2, 1.2)
#plt.grid(True)
plt.legend() 
plt.show()
```

```python
# Example 3

import numpy as np
import matplotlib.pyplot as plt
from time import sleep

v_perp = 1; w = 1
t = np.linspace(0, 50, 500)

x = (v_perp / w) * np.sin(w * t)
y = (v_perp / w) * np.cos(w * t)
z = 2*t

fig = plt.figure()
ax = fig.add_subplot(projection='3d')    
ax.grid(False)
ax.plot(x, y, z, '+b')
#ax.set_xlim(-1.5, 1.5); ax.set_ylim(-1.1, 1.1); ax.set_zlim(0, 100)
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z')
ax.view_init(elev=30, azim=-30)
plt.show()
```
