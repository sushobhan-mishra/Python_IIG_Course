```python
import numpy as np
import matplotlib.pyplot as plt
# from matplotlib.colors import LightSource

#constants
r =3000
phi = np.radians(np.linspace(0,180,100))
theta = np.radians(np.linspace(-90,90,100))

phi_m,theta_m = np.meshgrid(phi,theta)

# equation
x = r*np.sin(phi_m)*np.cos(theta_m)
y = r*np.sin(phi_m)*np.sin(theta_m)
z = r*np.cos(phi_m)

# ls = LightSource(azdeg=0,altdeg=45)

# rgb = ls.shade(z, cmap=plt.cm.Blues,vert_exag=0.1,blend_mode='soft')


# spacecraft orbit
df.rename(columns={'SPACECRAFT_X_MSO_(@_1_)_km': 'SPACECRAFT_X'}, inplace=True)
df.rename(columns={'SPACECRAFT_Y_MSO_(@_2_)_km': 'SPACECRAFT_Y'}, inplace=True)
df.rename(columns={'SPACECRAFT_Z_MSO_(@_3_)_km': 'SPACECRAFT_Z'}, inplace=True)

SPACECRAFT_X = df.iloc[:,15]
SPACECRAFT_Y = df.iloc[:,16]
SPACECRAFT_Z = df.iloc[:,17]

SPACECRAFT_X = SPACECRAFT_X.mask(SPACECRAFT_X < -3e8, np.nan)
SPACECRAFT_Y = SPACECRAFT_Y.mask(SPACECRAFT_Y < -3e8, np.nan)
SPACECRAFT_Z = SPACECRAFT_Z.mask(SPACECRAFT_Z< -3e8, np.nan)

t_utc = df.iloc[:,0]
t_utc = pd.to_datetime(t_utc)

#plotting
fig = plt.figure(layout='constrained')
ax = fig.add_subplot(projection='3d')    
# ax.grid(False)
ax.plot_surface(x,y,z,linewidth=1,facecolor='orange',alpha=1)
ax.plot_surface(-x,-y,-z,linewidth=1,facecolor='black',alpha=1)
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z')

#plotting orbit
ax.plot(SPACECRAFT_X, SPACECRAFT_Y, SPACECRAFT_Z)
ax.axis('equal')

plt.show()
```
