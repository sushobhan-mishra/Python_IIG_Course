Type Casting

```python
S = set([1, 3, 5, 1, 5, 7]) #list
print(S)
S = set((1, 2, 2, 3, 5, 6, 3, 4)) #tuple
print(S)
S = set("Apple") #string
print(S)
S = set("AappleE")  #string
print(S)
S = set(range(1,4)) #range
print(S)
S = set({'a':1,'b':2})  #dict
print(S)
```

Common methods

```python
s = {1,2,3,4,5}
#add
s.add(111)
print(s)

#update
s.update([64,222,333])
print(s)

#remove (Raise error when no element)
s.remove(111)
print(s)

#discard (Do not raise error when no element)
s.discard(77)
print(s)
s.discard(64)
print(s)
```

```python
s = {64,1,2,3,4,5, 333, 222, 111}
#pop (remove arbitrary element of sets)
print(s.pop())

#clear
print(s.clear())
print(s)
```

```python
s = {64,1,2,3,4,5, 333, 222, 111}
copy_s = s.copy()
print(copy_s)
```

Set Operation Methods

```python
a = {1, 2, 3, 4} 
b = {3, 4, 5, 6}
c = {5, 6, 7, 8}
d = {1, 2, 3, 4, 5, 6}

print(2 in a)
print(5 in a)
print(5 not in a)
print(1 not in a)
```

```python
a = {1, 2, 3, 4} 
b = {3, 4, 5, 6}
c = {5, 6, 7, 8}
d = {1, 2, 3, 4, 5, 6}
#disjoint (no common elements in both sets)
print(a.isdisjoint(c))
print(a.isdisjoint(b))

#subset (all elements in a differnt set)
print(a.issubset(d))
print(a.issubset(b))

#superset (opposite of subset)
print(d.issuperset(a))
print(a.issuperset(b))
```

```python
a = {1, 2, 3, 4} 
b = {3, 4, 5, 6}
c = {5, 6, 7, 8}
d = {1, 2, 3, 4, 5, 6}
#union
print(a.union(b))
print(a.union(b, c))
print(a | b)
print(a | b | c)

#intersection
print(a.intersection(b))
print(b.intersection(c, d))
print(a & b)

#difference
print( a.difference(b) )
print(a - b)

#symmetric_difference
print( a.symmetric_difference(b))
print(a ^ b)
```

```python
a = {1, 2, 3, 4} 
b = {3, 4, 5, 6}
c = {5, 6, 7, 8}
d = {1, 2, 3, 4, 5, 6}
#update |=
#intersection_update
print(a.intersection(b))
print(a)
print(b)
print(a.intersection_update(b))
print(a)
print(b)

#difference_update
print(a.difference_update(b))
print(a)
print(b)

#symmetric_difference_update
print(a.symmetric_difference_update(b))
print(a)
print(b)
```
