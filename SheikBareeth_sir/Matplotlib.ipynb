Oject Oreinted Plotting

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()

ax1 = ax.twinx()     #for sharing x with y and z

x,y,z = [1,2,3,4,5],[1,4,9,16,25],[1,8,27,64,125]

ax.plot(x,y,'b:',label = '$x^2$')
ax1.plot(x,z,'r--',label = '$x^3$')

line1,label1 = ax.get_legend_handles_labels()
line2,label2 = ax1.get_legend_handles_labels()

ax.set_title('Plotting two variables')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax1.set_ylabel('Z')

ax.legend(line1 + line2,label1 + label2,loc = 'upper left')
plt.show()
```

More Customization

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

fig,ax = plt.subplots()

x,y = [1,2,3,4,5],[1,4,9,16,25]

ax.plot(x,y,'g:',label = '$X^2$')

line1,label1 = ax.get_legend_handles_labels()

ax.set_xlim(0,6)
ax.set_ylim(0,26)

ax.spines['bottom'].set_color('red')
ax.spines['left'].set_color('red')
ax.spines['top'].set_color('blue')
ax.spines['right'].set_color('blue')

ax.xaxis.set_major_locator(MultipleLocator(2))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(MultipleLocator(2.5))

ax.grid(which = 'major',alpha = 0.7)
ax.grid(which = 'minor',alpha = 0.3)

ax.set_title('Plotting two variables')
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.legend(line1,label1,loc = 'upper left')
plt.show()
```

Adding text and annotation to plots

```python
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots(figsize = (8,4),layout = 'constrained',facecolor = '0.8')
ax1 = ax.twinx()

t = np.linspace(0,10,500)
# x = Asin(wt) = Asin(2*pi*nu*t) = Asin(theta)
y = 5*np.sin(2*np.pi*0.2*t)
z = 2*np.cos(2*np.pi*0.1*t)

ax.plot(t,y,'b:',label = r'$Sin(\theta)$')
ax1.plot(t,z,'r--',label = r'$Cos(\omega t)$')

line1,label1 = ax.get_legend_handles_labels()
line2,label2 = ax1.get_legend_handles_labels()

ax.set_title('Plotting two variables')
ax.set_xlabel('Time')
ax.set_ylabel('Sine variation')
ax1.set_ylabel('Cosine variation')

ax.text(0, -3, 'Trigonometry\nfunctions', style='italic', bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})
ax.annotate('Sin function peak', xy = (np.arcsin(1)/(2*np.pi*0.2),5), xytext= (2.5, 4), arrowprops = dict(arrowstyle="->", facecolor  = 'red')) 

fig.legend(line1 + line2,label1 + label2,loc = 'outside upper right')
plt.show()
```

Adding infinite lines

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(-10,10,100)
sig = 1 / (1 + np.exp(-t))

fig,ax = plt.subplots(figsize = (7,4),facecolor = '0.6')

ax.axhline(y=0,color='black',linestyle='--')
ax.axhline(y=0.5,color='black',linestyle=':')
ax.axhline(y=1.0,color='black',linestyle='--')
ax.axvline(color='black')
ax.axline((0, 0.5), slope=0.25, color="black", linestyle=(0, (5, 5))) 

ax.plot(t, sig, linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")

ax.set(xlim=(-10, 10), xlabel="t") 

ax.set_title("Sigmoid function") 

ax.legend(fontsize=14) 
plt.show() 
```

Array of plots

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (6, 4), facecolor = "0.8", layout='constrained') 

fig.suptitle("Array of Plots", fontsize = 15) 

x = np.linspace(0, 200, 500) 
y1 = x**2
y2 = 3*np.sin(2*np.pi*x/50)
x3 = np.linspace(-200, 200, 1000)
y3 = np.exp(x3/50)

axs[0][0].plot(x, y1, linewidth = 1, label='$x^2$') 
axs[0][0].grid()     
axs[0][0].legend() 
axs[0][0].set_title('Square of variable')

axs[0][1].plot(x, y2, linewidth = 2, color = 'green', ls = '-') 
axs[0][1].axhline(y=0, color = 'grey', ls = ':') 
for xv in np.arange(0, 250, 50): 
    axs[0][1].axvline(x=xv,color='red', ls='--', lw = 0.7) 
axs[0][1].set_title('Sine wave generated') 
 
axs[1][0].plot(x3, y3, linewidth = 1, c='magenta') 
axs[1][0].axvline(x=0, color = 'black', ls = '--') 
axs[1][0].axhline(y=1, color = 'black', ls = ':') 
axs[1][0].text(8, 40, '(x=0)') 
axs[1][0].text(120, 4, '(y=1)') 
axs[1][0].set_title('Exponential function')

axs[1][1].plot(x, np.log(x) , linewidth = 1, c= 'orange') 
axs[1][1].set_title('Logarithmic function') 
axs[1][1].legend(['log(x)']) 

plt.show()
```
