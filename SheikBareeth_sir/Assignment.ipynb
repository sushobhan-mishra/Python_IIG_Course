```python
d_profile = {'name':'Sushobhan','age':21,'date of birth':'16_12_2004','gender':'male','qualification':'bsc','hometown':'bhadrak'}
d_profile.update({'father':'Satyajit Mishra','mother':'Eleena Senapati','income':1,'fav_subject':'Physics'},)

keys_to_remove = ["father", "mother", "income", "fav_subject"]

removed_items = []
for key in keys_to_remove:
    if key in d_profile:
        removed_items.append((key, d_profile.pop(key)))
        
print("Dictionary after removal:")
print(d_profile)
print("\nRemoved items list:")
print(removed_items)
removed_items = dict(removed_items)
print(removed_items)

```

```python
D ={}
for i in range(4):
    roll = input(f'Enter roll no. {i+1}:')
    name = input(f'Enter name {i+1}:')
    clas = input(f'Enter class of {name}:')
    age = input(f'Enter age of {name}:')
    height = input(f'Enter height of {name}:')
    marks = int(input(f'Enter marks of {name}:'))
    if marks > 100:
        a = 'wrong input'
    elif marks < 40:
        a = 'fail'
    elif marks >= 40:
        a = 'pass'      
    details = {'name': name,'class': clas ,'age': age,'height' : height,'marks' : marks,'Result':a}
    D[roll] = details
print(D) 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for 24 kHz spectrum plot
f = np.linspace(1, 50, 500)  # frequency (kHz)
amp = np.exp(-((f - 10) / 8) ** 2) * np.sin(2 * np.pi * f / 12) + 0.15 * np.random.randn(len(f))

plt.figure(figsize=(8, 4))
plt.plot(f, amp)
plt.title("24 kHz Spectrum (sample plot)")
plt.xlabel("Frequency (kHz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for a 24 MHz spectrum plot
f = np.linspace(1, 200, 2000)  # Frequency in MHz
amp = np.exp(-((f - 24) / 6)**2) * np.sin(2*np.pi*f/40) + 0.05*np.random.randn(len(f))

plt.figure(figsize=(8, 4))
plt.plot(f, amp)
plt.title("24 MHz Spectrum (sample plot)")
plt.xlabel("Frequency (MHz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
