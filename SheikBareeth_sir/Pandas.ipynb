isna() used to find the null values in the dataset. It returns a boolean value True for null values and False for non-null values.

```python
ndata = {"country": ["India", "China", "Brazil", "Japan", "Germany",  
     "Italy", "Russia", "Spain", "Peru", "Kenya"],  
     "year": [2002, 2005, 2004, 2006, 2001, 2003, 2002, 2005, 2004,2006 ],
     "population": [2.5, 2.8, None, 0.5, 1.2, 2.1, 1.6, 1.9, 2.0, None]}

dF = pd.DataFrame(ndata)
print(dF.isna())
```

any(axis=1) returns true for a rows having at least one missing value

```python
nF = dF[dF.isna().any(axis=1)]
print(nF)
```

fillna for fill something in null data

```python
print(dF.fillna('NotAvailable'))
```

```python
Data = {"country": ["India", "China", "Brazil", "Japan", "Germany",  
     "Italy", "Russia", "Spain", "Peru", "Kenya"],  
     "year": [2002, 2005, 2004, 2006, 2001, 2003, 2002, 2005, 2004,2006 ],
     "population": [2.5, 2.8, None, 0.5, 1.2, 2.1, 1.6, 1.9, 2.0, None]}

dF = pd.DataFrame(Data)

dF = dF.set_index("year")

dF = dF.sort_index()

print(dF.loc[2003:2006, : ]) 
```

```python
Data = {"country": ["India", "China", "Brazil", "Japan", "Germany",  
     "Italy", "Russia", "Spain", "Peru", "Kenya"],  
     "year": [2002, 2005, 2004, 2006, 2001, 2003, 2002, 2005, 2004,2006 ],
     "population": [2.5, 2.8, None, 0.5, 1.2, 2.1, 1.6, 1.9, 2.0, None]}

dF = pd.DataFrame(Data)

print(dF.iloc[0 : 4 , :]) 
 # slicing rows(0 to 3 like index) and columns(all)
```

```python
print(dF.iloc[4:8, 0:2]) 
# slicing rows(4 to 7 like index) and columns(0 to 1 like index)
```

Filtering Data

```python
data = {'Name': ['Raghav', 'Sourabh', 'Veer', 'Iyer', 'Gupta', 'Kumar', 'Sharma'], 
               'Age': [25, 32, 45, 19, 27, 18, 29], 
             'Score': [85, 90, 78, 95, 35, 48, 71]} 
dF = pd.DataFrame(data)

dF_filtered = dF[dF['Age'] > 30]

print(dF_filtered)
```

```python
new_DF = dF[(dF["Age"] < 30) | (dF["Score"] >50)] 
print(new_DF)
```

```python
new_DF = dF[(dF["Age"] < 30) & (dF["Score"] >70)] 
print(new_DF)
```

```python
new_DF = dF.loc[dF['Age'] > 30, ['Name', 'Score']] 
print(new_DF)
```

```python
new_DF = dF[~(dF['Age'] > 26)] 
print(new_DF)
```

If we have the values of the items to filter, we can use isin()

```python
new_DF = dF[dF['Age'].isin([18, 19])] 
print(new_DF)            
```
