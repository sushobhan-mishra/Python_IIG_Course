```python
'''2. Using list comprehension, create a list, where each element is the sum of the current integer and its previous integer. Use range(20) as x-axis and plot the generated list as y-axis.
(a) Set the title for the plot.
(b) Set x label and y label of your choice.
(c) Set line width to 3.
(d) Experiment with at least 3 different combinations of color, marker, size,and font, and choose your best one. '''

import matplotlib.pyplot as plt

x = list(range(20))
y = [i + (i - 1) for i in x]

plt.plot(x, y, color='navy', marker='D', markersize=7, linewidth=3)

plt.title("Sum of current and previous integer", fontname="DejaVu Sans", fontsize=14)
plt.xlabel("x (0 to 19)", fontname="DejaVu Sans", fontsize=12)
plt.ylabel("y = x + (x-1)", fontname="DejaVu Sans", fontsize=12)

# plt.xticks(x)
plt.grid(True, alpha=0.7)
plt.show()
```

```python
'''3.  Using NumPy, generate two arrays of random values: one in the range 5 to 10  
     (for x-axis) and another in the range 1 to 5 (for y-axis). Plot these values using a  
      scatter plot in Matplotlib.  
        (a) Set x label, y label, plot title. 
        (b) Change Marker shape, marker size and color 
        (c) change transparency for 3 different "alpha".'''

import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(5, 10, size=30)    #unifrom for float
y = np.random.uniform(1, 5, size=30)  

plt.xlabel("X (range 5 to 10)")
plt.ylabel("Y (range 1 to 5)")
plt.title("Random Scatter Plot using NumPy")

plt.scatter(x, y, marker="o", s=10, c="red", alpha=0.9, label="alpha=0.9")
plt.scatter(x, y, marker="s", s=40, c="green", alpha=0.5, label="alpha=0.5")
plt.scatter(x, y, marker="^", s=80, c="blue", alpha=0.2, label="alpha=0.2")

plt.legend(loc = 'best')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
'''4.  Create a list of 30 runs (values between 0 to 100 generated randomly)  
      representing scores by Virat Kohli in 30 matches. Create histogram, 
      (a) With default bin edges 
      (b) With Bin edges of your choice'''

import numpy as np
import matplotlib.pyplot as plt

# .tolist() works for 1D 2D or 3D array to covert to list
# But list() works only for 1D array
runs = np.random.randint(0, 101, size=30).tolist()   # randint for int

plt.hist(runs)
plt.title("Virat Kohli Scores (Default bins)")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()

bin_edges = [0, 20, 40, 60, 80, 100]
plt.hist(runs, bins=bin_edges, edgecolor="black")
plt.title("Virat Kohli Scores(Custom bins)")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
'''7. Plot a Sine wave using subplots with only one axes. 
  (a) Set axes title, X label, Y label. 
  (b) Create legend for the plot. 
  (c) Place the legend in 4 different locations. '''

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

fig,ax = plt.subplots(nrows= 1,ncols= 1)

ax.plot(x, y, label='sin(x)', linewidth=2)

ax.set_title('Sine Wave')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')

locations = ['upper right', 'lower left', 'upper left', 'lower right']

for loc in locations:
    ax.legend(loc=loc)
    print(f"Legend placed at: {loc}")
plt.show()
```

```python
'''9. Create a figure containing 2x2 subplots. Plot in all four axes with Figure title
    and axes titles of your Choice. Plotting variables are of your choice. Use
    legends, texts.'''

import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2 * np.pi, 1000)

sine = np.sin(t)
cosine = np.cos(t)

t_log = np.linspace(0.01, 2 * np.pi, 1000)
log_t = np.log(t_log)

# unit circle parametric
theta = np.linspace(0, 2 * np.pi, 400)
cx, cy = np.cos(theta), np.sin(theta)

# plot
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8), layout='constrained')
fig.suptitle('Waveforms, Log, and Circle', fontsize=16)

# Sine
axs[0, 0].plot(t, sine, label='sin(t)', color='tab:blue')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_xlabel('t')
axs[0, 0].set_ylabel('sin(t)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Cosine
axs[0, 1].plot(t, cosine, label='cos(t)', color='tab:orange')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].set_xlabel('t')
axs[0, 1].set_ylabel('cos(t)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Log plot (natural logarithm)
axs[1, 0].plot(t_log, log_t, label='ln(t)', color='tab:green')
axs[1, 0].set_title('Natural Log of t')
axs[1, 0].set_xlabel('t')
axs[1, 0].set_ylabel('ln(t)')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Circle (parametric)
axs[1, 1].plot(cx, cy, label='circle (r=1)', color='tab:red')
axs[1, 1].set_title('Unit Circle')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].axis('equal')
axs[1, 1].legend()
axs[1, 1].grid(True)

plt.show()
```

```python
'''10. Create a 2x1 subplot. Share "X" axis. "Y" variable for bottom axes is Sin(2*  
      Pi*t/2). For top axes, Sin(2* Pi*t/4). Set titles, labels, legends etc (all that you  
      have learned). (Hint. Use NumPy Linspace for t.)'''  

import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 8, 1000)

y_top = np.sin(2 * np.pi * t / 4)   
y_bottom = np.sin(2 * np.pi * t / 2) 

fig, axs = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(9, 6), layout='constrained')
fig.suptitle('Two Sine Waves', fontsize=14)

axs[0].plot(t, y_top, label='sin(2π t / 4)', color='blue')
axs[0].set_title('Lower Frequency')
axs[0].set_ylabel('Amplitude')
axs[0].legend(loc='upper right')
axs[0].grid(True)

axs[1].plot(t, y_bottom, label='sin(2π t / 2)', color='orange')
axs[1].set_title('Higher Frequency')
axs[1].set_xlabel('t')
axs[1].set_ylabel('Amplitude')
axs[1].legend(loc='upper right')
axs[1].grid(True)

plt.show()
```
