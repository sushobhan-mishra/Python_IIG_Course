Creating dictionary

```python
D = {'A':1,'B':2,'C':3}
print(D)
```

```python
D = dict(A=1,B=2,C=3)
print(D)
```

```python
D = dict([('A',1),('B',2),('C',3)])
print(D)
```

```python
Alphabets = ['A','B','C']
positions = [1,2,3]
D = dict(zip(Alphabets,positions))
print(D)
```

Dictionary Methods

GETTING DATA

```python
D = {'A':1,'B':2,'C':3,'D':4}
print(D.get('A'))
print(D.get("A", 22))  
print(D.get("G"))                 
print(D.get("G", 111)) 
```

```python
print(D.keys())
print(list(D.keys()))
print(D.values())
print(list(D.values()))
print(D.items())
```

ADDING AND UPDATING DATA

```python
D = {'A':1,'B':2,'C':3,'D':4}
D.update({'C':33})
print(D)
D.update({'B':222,'E':5})
print(D)
```

```python
D = {'A':1,'B':2,'C':3,'D':4}
print(D.setdefault('A',2))
print(D)
print(D.setdefault('E',5))
print(D)
```

REMOVING DATA

```python
D = {'A':1,'B':2,'C':3,'D':4}
print(D.pop('B'))
print(D)
print(D.popitem())            #removes last item
print(D)
D.clear()
print(D)
```

```python
d = {"a" : [1, 2, 3]} 
copy_d = d.copy()
copy_d['a'].append(4)
print(copy_d)
print(d)
```

```python
d = {"a" : [1, 2, 3]} 
copy_d = d.copy()
copy_d['a'] = [111,444]
print(copy_d)
print(d)
```

Creating dictionary

```python
D = dict.fromkeys(['a','b','c'],5)
print(D)
```

```python
D = { i : i for i in [1, 2, 3, 4]}
print(D)
D = { f"{i}" : i**2 for i in range(1, 4)}
print(D)

s = [10, 20, 30, 40, 50]
D = { f"{i//10}" : i for i in range(100) if i in s}
print(D)
```
