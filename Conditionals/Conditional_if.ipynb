```python
X = input('Enter something:')
if X.isnumeric():
    print(f'{X} is a number')
elif X.isalpha():
    print(f'{X} is a string')
elif int(X) < 0:
    print(f'{X} is negative')    
```

```python
X = int(input('Enter a number:'))
if X > 0:
    print(f'{X} is a positive number')
elif X < 0:
    print(f'{X} is a negative number') 
else:
    print(f'{X} is zero')     
```

```python
X = int(input('Enter a number:'))
if X % 2 == 0:
    print(f'{X} is even.')
else:
    print(f'{X} is odd.')    
```

```python
a,b,c = input('Enter the three number with space :').split(' ')

if a >= b and a >= c:
    print(f'{a} is greater ')
elif b >= a and b >= c:
    print(f'{b} is greater')
else:
    print(f'{c} is greater')        
```

```python
a,b,c,d = input('Enter the three number with space :').split(' ')

if a >= b and a >= c and a >= d:
    print(f'{a} is greater ')
elif b >= a and b >= c and b >= d:
    print(f'{b} is greater')
elif c >= a and c >= b and c >= d:
    print(f'{c} is greater')
else:
    print(f'{d} is greater')    
```
